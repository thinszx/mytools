```python
SERVER = "127.0.0.1"
PORT = 1234
USERNAME = "username"
PASSWORD = "password"
SSL = False
```

```python
import os

EXPORT_DIR = "/mnt/d/Documents/tmp"
EXPORT_FORMAT = "CVAT for video 1.1"
EXPORT_INCLUDE_IMAGES = False

if not os.path.exists(EXPORT_DIR):
    os.makedirs(EXPORT_DIR)
```

```python
from cvat_sdk import make_client
from cvat_sdk.core.helpers import TqdmProgressReporter # https://github.com/opencv/cvat/blob/develop/cvat-cli/src/cvat_cli/cli.py#L12
import tqdm

if SSL:
    HOST = f"https://{SERVER}"
else:
    HOST = f"http://{SERVER}"

with make_client(host=HOST, port=F"{PORT}", credentials=(f"{USERNAME}", f"{PASSWORD}")) as client:
    tasks = client.tasks.list()
    # jobs = client.jobs.list()
    for task in tqdm.tqdm(tasks):
        # pbar = TqdmProgressReporter(
        #         tqdm.tqdm_notebook(unit_scale=True, unit="B", unit_divisor=1024, desc=task.name)
        #     )
        if task.id > 60:
            continue
        try:
            pbar = None
            client.tasks.retrieve(obj_id=task.id).export_dataset(
                    format_name=EXPORT_FORMAT,
                    filename=os.path.join(EXPORT_DIR, f"{task.name}.zip"),
                    pbar=pbar,
                    status_check_period=2,
                    include_images=EXPORT_INCLUDE_IMAGES,
                )
        except:
            pass
```
